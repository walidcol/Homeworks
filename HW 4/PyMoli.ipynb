```python
df = pd.read_csv(file)
df.head()
```

```python
# total number of players

players = df["Gender"].count()
players
```

```python
# PURCHASING ANALYSIS (TOTAL)

# Number of Unique Items
unique_items = df["Item Name"].count()

# Average Purchase Price
average_price = df["Price"].mean()

# Total Number of Purchases
total_purchases = df["Purchase ID"].count()

# Total Revenue
total_revenue = df["Price"].sum()

purchase_analysis = pd.DataFrame ({
    
    "Number of Unique Items" : [unique_items],
    "Average Purchase Price" : [average_price],
    "Total Number of Purchases" : [total_purchases],
    "Total Revenue" : [total_revenue]
    
    })

purchase_analysis
```

```python
# GENDER DEMOGRAPHICS

# Percentage and Count of Male Players
players_by_gender = df["Gender"].value_counts()

#puts the data into a table/dataframe
players_by_gender_df = pd.DataFrame(players_by_gender)

players_by_gender_df
```

```python
# to get percentage of genders

percent_genders = (players_by_gender / players) * 100
percent_genders_df = pd.DataFrame(percent_genders)
percent_genders_df

#making table look nicer (adding percent of players and renaming gender)

#new_table = pd.DataFrame(percent_genders_df)
new_table = pd.DataFrame(players_by_gender_df)
new_table

#making table look nicer (adding percent of players and renaming gender)
new_table = new_table.rename(columns ={ "Gender" : "Total Count"})

new_table["Percent of PLayers"] = percent_genders
new_table
```

```python
# Purchasing Analysis (Gender)

#purchase count by gender

all_gender = df.groupby("Gender")

player_count = all_gender["SN"].count()
player_count_df = pd.DataFrame(player_count)

# Average Purchase Price by gender
average_price = all_gender["Price"].mean()
average_price_df = pd.DataFrame(average_price)
#average_price_df

# Total Purchase Value by gender

total_purchase = all_gender["Price"].sum()

# Average Purchase Total per Person by Gender

avg_purchase_by_gender = total_purchase / player_count

# Add 'Average Purchase Price' to table

player_count_df["Average Purchase Price"] = average_price

# Rename 'SN' column to 'Purchase Count'

new_player_count = player_count_df.rename(columns = {"SN" : "Purchase Count"})

# Add 'Normalized Totals' to table

new_player_count["Total Purchase Price"] = total_purchase

new_player_count_df = pd.DataFrame(new_player_count)

# Add 'Total Purchase Price' to table

new_player_count_df["Avg Total by Gender"] = avg_purchase_by_gender

total_gender_purchase_df = pd.DataFrame(new_player_count_df)

total_gender_purchase_df



```

```python
# The below each broken into bins of 4 years

bins = [0,10,15,20,25,30,35,40,100]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40<"]

df["Age Demographics"] = pd.cut(df["Age"], bins, labels = group_names)
demo_group = df.groupby("Age Demographics")
df.head()
```

```python
# TOP SPENDERS

#total purchase count per SN

SN_count = df.groupby("SN")["Price"].count()
SN_count_df = pd.DataFrame(SN_count)

# avg purchase price per SN

SN_average = df.groupby("SN")["Price"].mean()

#total purhcase price per SN

SN_total = df.groupby("SN")["Price"].sum()


#add data to table

SN_count_df["Average Price"] = SN_average

SN_count_df["Total Purchase Value"] = SN_total

SN_count_df.rename(columns = {"Price" : "Total Purchase Count"})


final_table_df = SN_count_df.sort_values(by = ["Total Purchase Value"], ascending = False)

final_table_df.head()






```

```python
# MOST POPULAR ITEMS

#Identify the 5 most popular items by purchase count
#then list (in a table)

#item id
item_id = df.groupby(df["Item ID"])

items = item_id["Item ID"].unique()

#item name
item_name = item_id["Item Name"].unique()

item_purchase_count = item_id["Age"].count()

item_price = item_id["Price"].unique()

item_total_value = item_id["Price"].sum()

table = {"Item ID" : items, "Item Name" : item_name, "Item Price" : item_price, "Item Count" : item_purchase_count, "Total Purchase" : item_total_value }

table_df = pd.DataFrame(table)

table_df = table_df.set_index("Item ID")

table_df = table_df.sort_values("Item Count", ascending = False)

table_df = table_df[["Item Name", "Item Count", "Item Price", "Total Purchase"]]

table_df.head()


```

```python
# MOST PROFITABE ITEMS



Profit_Table = table_df.sort_values("Total Purchase", ascending = False)
Profit_Table.head()

```
